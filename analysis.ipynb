```python
file = open("my.log")
```

```python
lines = file.readlines()
```

```python
import re
pattern = "INFO:flwr:fit progress: \(([1-9]\d*), ([1-9]\d*\.\d*|0\.\d*[1-9]\d*), \{'accuracy': ([1-9]\d*\.\d*|0\.\d*[1-9]\d*)\}"
data = []
for line in lines:
    matchObj = re.match(pattern, line)
    if matchObj != None:
        data.append([int(matchObj.group(1)), 
                     float(matchObj.group(2)), 
                     float(matchObj.group(3))
                     ])
```

```python
import pandas as pd

df = pd.DataFrame(data, columns=["round", "loss", "accuracy"])
df
```

```python
df.plot(x="round", y="accuracy")
```

```python
df.plot(x="round", y="loss")
```
